```python
df2 = pd.DataFrame()
df2['assay'] = toxcast_assay.Name
df2['sample_size'] = toxcast_assay.Chemicals
df2['active'] = toxcast_assay["Active%"]
df2.index = df2['assay']
df2 = df2.drop("assay", axis=1).T
```

```python
active_df = pd.DataFrame()
for assay in moleculnet_toxcast.columns[1:] : 
  values = moleculnet_toxcast[assay].value_counts()
  sample_size = int(sum(values))
  active_p = round( values[1]/sample_size ,4) * 100
  active_df[assay] = [sample_size,active_p]
```

```python
df_949_info = df2 # z-score로 필터링 된것
df_molculnet_info = active_df #z-score로 필터링 되지 않음.
```
